```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, integrate, diff, Eq, solve, exp
```

# Норма и разстояние
## Намиране на най-добро среднокрадратично приближение

### Задача 1

Да се намери полиномът на най-добро средноквадратично приближение от втора степен в интервала $[0, 1.5]$ за функцията $f(x) = e^x$ при тегло $\mu(x) = 1$.

```python
a, b, c, x = symbols("a, b, c, x")

def f(x):
    return exp(x)

def g(x):
    return a*x**2+b*x+c

def phi(a, b, c, x):
    return integrate((f(x)-g(x))**2, (x, 0, 1.5))

equations = [Eq(diff(phi(a, b, c, x), a), 0), Eq(diff(phi(a, b, c, x), b), 0), Eq(diff(phi(a, b, c, x), c), 0)]

sol = solve(equations)
x_axis = np.linspace(0, 1.5, 100)
plt.plot(x_axis, np.exp(x_axis), x_axis, g(sol[a], sol[b], sol[c], x_axis))
```
